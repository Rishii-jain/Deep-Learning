```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
import pickle
from keras import Sequential
from keras.datasets import mnist
```

```python
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(28, kernel_size=(3,3), padding = 'valid',strides=(2,2), activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(28, kernel_size=(3,3), padding = 'valid', strides=(2,2),activation = 'relu'))
model.add(Conv2D(28, kernel_size=(3,3), padding = 'valid', strides=(2,2),activation = 'relu'))

model.add(Flatten())

model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation = 'softmax'))
```

```python
model.summary()
```

```python
model = Sequential()

model.add(Conv2D(28, kernel_size=(3,3), padding = 'same', activation = 'relu', input_shape = (28,28,1)))
model.add(Conv2D(28, kernel_size=(3,3), padding = 'same', activation = 'relu'))
model.add(Conv2D(28, kernel_size=(3,3), padding = 'same', activation = 'relu'))

model.add(Flatten())

model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation = 'softmax'))
```
